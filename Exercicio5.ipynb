```python
# Exercicio 5 - letra A

beta = 2;
lambda1 = 1/beta;
print(lambda1, "chegadas/minuto")
```

```python
# Exercicio 5 - letra B

import numpy as np
import matplotlib.pyplot as plt

N = 100000
lambda1 = 0.5  # Se a média é 2 minutos, então lambda = 1/2 = 0.5
x = np.random.uniform(0, 1, N) # Gera N amostras uniformes

Xexp = -np.log(x) / lambda1 # Converte para Exponencial: X = -ln(x)/lambda

# Probabilidade simulada de X < 1
count = np.sum(Xexp < 1)
prob_simulada = count / N

# Probabilidade teórica de X < 1
prob_teorica = 1 - np.exp(-lambda1 * 1)

print(f"Prob. simulada de X < 1: {prob_simulada:.4f}")
print(f"Prob. teórica  de X < 1: {prob_teorica:.4f}")

# Plot do histograma das amostras
plt.hist(Xexp, bins=100, density=True, alpha=0.6, label='Simulado')

# Curva da densidade teórica
x = np.linspace(0, 10, 200)
f_teor = lambda1 * np.exp(-lambda1 * x)
plt.plot(x, f_teor, 'r-', label='Teórico (PDF)')
plt.title('Distribuição Exponencial (média=2, λ=0.5, Probabilidade < 1)')
plt.xlabel('Tempo (min)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

```python
# Exercicio 5 - letra C

import numpy as np
import matplotlib.pyplot as plt

N = 100000
lambda1 = 0.5  # Se a média é 2 minutos, então lambda = 1/2 = 0.5
x = np.random.uniform(0, 1, N) # Gera N amostras uniformes

Xexp = -np.log(x) / lambda1 # Converte para Exponencial: X = -ln(x)/lambda

# Probabilidade simulada de X > 4
count = np.sum(Xexp > 4)
prob_simulada = count / N

# Probabilidade teórica de X > 4
prob_teorica = np.exp(-lambda1 * 4)

print(f"Prob. simulada de X > 4: {prob_simulada:.4f}")
print(f"Prob. teórica  de X > 4: {prob_teorica:.4f}")

# Plot do histograma das amostras
plt.hist(Xexp, bins=100, density=True, alpha=0.6, label='Simulado')

# Curva da densidade teórica
x = np.linspace(0, 10, 200)
f_teor = lambda1 * np.exp(-lambda1 * x)
plt.plot(x, f_teor, 'r-', label='Teórico (PDF)')
plt.title('Distribuição Exponencial (média=2, λ=0.5, Probabilidade > 4))')
plt.xlabel('Tempo (min)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

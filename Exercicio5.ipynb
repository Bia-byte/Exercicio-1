```python
# Exercicio 5 - letra A

beta = 2;            # média da distribuição exponencial
lambda1 = 1/beta;    # número médio de requisições
print(lambda1, "chegadas/minuto")
```

```python
# Exercicio 5 - letra B

import numpy as np
import matplotlib.pyplot as plt

N = 100000     # Número de amostras
lambda1 = 0.5  # Se a média é 2 minutos, então lambda = 1/2 = 0.5
x = np.random.uniform(0, 1, N) # Gera N amostras uniformes
value = 1      # valor de x (número de tentativas)
count=0        # Contador para somar as probabilidades menor que 1
Xexp = -np.log(x) / lambda1 # Converte para Exponencial: X = -ln(x)/lambda

# Probabilidade simulada de X < 1
for expvalue in Xexp:
    if expvalue <= value:  #enquanto o valor gerado for menor ou igual ao valor de x
        count=count+1    # incrementa o contador
prob_simulada=count/N    # soma das probabilidades menores do que 1 dividido pelo total de amostras

# Probabilidade teórica de X < 1
prob_teorica = 1- np.exp(-lambda1 * value)

print(f"Prob. simulada de X < 1: {prob_simulada:.4f}")
print(f"Prob. teórica  de X < 1: {prob_teorica:.4f}")

# Plot do histograma das amostras
plt.hist(Xexp, bins=100, density=True, alpha=0.6, label='Simulado')
# Curva da densidade teórica
x = np.linspace(0, 10, 200)
f_teor = lambda1 * np.exp(-lambda1 * x)
plt.plot(x, f_teor, 'r-', label='Teórico (PDF)')
plt.title('Distribuição Exponencial (média=2, λ=0.5, Probabilidade < 1))')
plt.xlabel('Tempo (min)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

```python
# Exercicio 5 - letra C

import numpy as np
import matplotlib.pyplot as plt

N = 100000     # Número de amostras
lambda1 = 0.5  # Se a média é 2 minutos, então lambda = 1/2 = 0.5
x = np.random.uniform(0, 1, N) # Gera N amostras uniformes
value = 4      # valor de x (número de tentativas)
count=0        # Contador para somar as probabilidades menores que 4
Xexp = -np.log(x) / lambda1 # Converte para Exponencial: X = -ln(x)/lambda

# Probabilidade simulada de X > 4
for expvalue in Xexp:
    if expvalue > value:  #enquanto o valor gerado for maior ao valor de x
        count=count+1    # incrementa o contador
prob_simulada=count/N    # soma das probabilidades menores do que 4 dividido pelo total de amostras

# Probabilidade teórica de X > 4
prob_teorica = 1 - (1- np.exp(-lambda1 * value))

print(f"Prob. simulada de X > 4: {prob_simulada:.4f}")
print(f"Prob. teórica  de X > 4: {prob_teorica:.4f}")

# Plot do histograma das amostras
plt.hist(Xexp, bins=100, density=True, alpha=0.6, label='Simulado')
# Curva da densidade teórica
x = np.linspace(0, 10, 200)
f_teor = lambda1 * np.exp(-lambda1 * x)
plt.plot(x, f_teor, 'r-', label='Teórico (PDF)')
plt.title('Distribuição Exponencial (média=2, λ=0.5, Probabilidade > 4))')
plt.xlabel('Tempo (min)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

```python
# Exercicio 7

import numpy as np
import matplotlib.pyplot as plt

N = 100000  # número de amostras a serem geradas

U = np.random.uniform(0, 1, N) # gera N valores aleatórios entre 0 e 1 (distribuição uniforme)

# Transformando as amostras uniformemente distribuídas via inversão para uma distribuição com pdf f(x) = 3x^2
# Transforma via inversa F^{-1}(u) = u^(1/3)
X = U**(1/3)  # Aplica a inversa da função de distribuição acumulada (CDF) para obter a distribuição desejada

# Curva da PDF analítica: f(x) = 3x^2, válida para 0 <= x <= 1
x_teorico = np.linspace(0, 1, 200)  # Gera 200 pontos igualmente espaçados no intervalo [0, 1]
f_teorico = 3*x_teorico**2         # Calcula os valores da PDF analítica para os pontos gerados

# Plotagem do histograma normalizado e da pdf analítica
plt.hist(X, bins=50, range=(0,1), density=True, alpha=0.6, edgecolor='k', label='Histograma (simulado)')
plt.plot(x_teorico, f_teorico, 'r-', linewidth=2, label='PDF analítica: $3x^2$')
plt.title('Distribuição com pdf $f(x)=3x^2$ (0 <= x <= 1) via Inversão')
plt.xlabel('x')
plt.ylabel('Densidade de probabilidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Exercicio 7

import numpy as np
import matplotlib.pyplot as plt

N = 100000

U = np.random.uniform(0, 1, N) # Gera N amostras U uniformes

X = U**(1/3) # Transforma via inversa F^{-1}(u) = u^(1/3)

# Curva da pdf analítica: f(x) = 3x^2, 0<=x<=1
x_teorico = np.linspace(0,1,200)
f_teorico = 3*x_teorico**2

# Plotagem do histograma normalizado e da pdf analítica
plt.hist(X, bins=50, range=(0,1), density=True, alpha=0.6, edgecolor='k', label='Histograma (simulado)')
plt.plot(x_teorico, f_teorico, 'r-', linewidth=2, label='PDF analítica: $3x^2$')
plt.title('Distribuição com pdf $f(x)=3x^2$ (0 <= x <= 1) via Inversão')
plt.xlabel('x')
plt.ylabel('Densidade de probabilidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
